```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer = pd.read_csv("customer.csv")
```

```python
products = pd.read_csv("Products.csv")
```

```python
sales = pd.read_csv("sales.csv")
```

```python
inventory = pd.read_csv("Inventory.csv")
```

```python
new_df = customer.merge(sales, on = "customer_id")
```

```python
new_df
```

```python
df = new_df.merge(products, on = "product_id")
```

```python
df["total_revenue"] = df["price"] * df["quantity"] 
df
```

```python
product_names =  df['name_y'].value_counts().index
```

```python
product_values =  df['name_y'].value_counts().values
```

```python
plt.pie(product_values, labels = product_names, autopct="%1.2f%%")
```

```python
category_names =  df['category'].value_counts().index
```

```python
category_values =  df['category'].value_counts().values
```

```python
plt.pie(category_values, labels = category_names, autopct="%1.2f%%")
```

```python
plt.bar("category","price", color ='blue',width = 0.4, data = df)
plt.xlabel("category")
plt.ylabel("price")
plt.title("Price for different category")
plt.show()
```

```python
age_names =  df['age'].value_counts().index
age_values =  df['age'].value_counts().index
plt.pie(age_values, labels = age_names, autopct="%1.2f%%")
```

```python
plt.bar("gender","total_revenue", color ='blue',width = 0.4, data = df)
plt.xlabel("gender")
plt.ylabel("total revenue")
plt.title("Total Revenue derived from each gender")
plt.show()
```

```python
plt.bar("age","total_revenue", color ='blue',width = 0.4, data = df)
plt.xlabel("age")
plt.ylabel("total revenue")
plt.title("Total Revenue derived from different age bracket")
plt.show()
```

```python
plt.bar("name_y","total_revenue", color ='blue',width = 0.4, data = df)
plt.xlabel("name_y")
plt.ylabel("total revenue")
plt.title("Total Revenue derived from each product name")
plt.show()
```
